```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df=pd.read_csv('Diwali sales Data.csv', encoding='unicode_escape') 
```

```python
df.shape #returns no. of rows and columns
```

```python
df.head(10) #top 10 rows are shown 
```

```python
df.info() #details of all columns
```

```python
df.drop(['Status','unnamed1'],axis=1,inplace=True )#axis deletes entire column and inplace savs the changes made
```

```python
df.info()
```

```python
pd.isnull(df) #check for null
```

```python
pd.isnull(df).sum() 
```

```python
df.dropna(inplace=True) #drop null values
```

```python
df.shape
```

```python
df['Amount']=df['Amount'].astype('int') #change type of column
```

```python
df[['Age','Amount']].describe() #retunrs mean sd of columns useful for seeing min n max values
```

```python
df.columns

```

```python
ax=sns.countplot(x='Gender',data=df) #plottig for gender in the data=df file
for bars in ax.containers:
    ax.bar_label(bars) #gives count on top of column
```

```python
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
#groups by gender returning the total sum of amounts arranged in ascending order
```

```python
sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',data=sales_gen)
#plotting it in barplot with x axis n y axis
```

Most of buyers are females with higher purchasing power

```python
sns.countplot(x='Age Group',data=df,hue='Gender') #gives a plot w.r.t  gender
```

```python
sns.countplot(x='Age Group',data=df)
```

```python
ax=sns.countplot(x='Age Group',data=df,hue='Gender') #plottig for gender in the data=df file
for bars in ax.containers:
    ax.bar_label(bars) #gives count on top of column
```

```python
sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',data=sales_age)
#plotting it in barplot with x axis n y axis
```

By looking at above 2 graphs we can understand that most purchases are made my women of age group 26-35 

```python
sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
#total no. of orders statewise of only  top 10  states . Hence head(10)
sns.set(rc={'figure.figsize':(25,5)}) #to prevent congestion in x axis width=25,height=5
sns.barplot(data=sales_state ,x='State',y='Orders')
```

```python
sales_state=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
#total no. of orders statewise of only  top 10  states . Hence head(10)
sns.set(rc={'figure.figsize':(25,5)}) #to prevent congestion in x axis width=25,height=5
sns.barplot(data=sales_state ,x='State',y='Amount')
```

From 2 graphs we can make out that though kerala has places max orders compared to haryana, Haryana has spent more amount in orders .

```python
ax=sns.countplot(data=df,x='Marital_Status')
sns.set(rc={'figure.figsize':(7,5)}) #last time's 25,5 will be considered here
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Marital_Status',y='Amount',data=sales_state,hue='Gender')
#plotting it in barplot with x axis n y axis
```

```python
#Now for Occupation
sns.set(rc={'figure.figsize':(20,5)})
ax=sns.countplot(data=df,x='Occupation')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_sector=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
#total no. of sector wise 

sns.barplot(data=sales_sector ,x='Occupation',y='Amount')
```

Mostly purchased by IT,Healthcare followed by Aviation 

```python
sns.set(rc={'figure.figsize':(25,5)})

ax=sns.countplot(data=df,x='Product_Category')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_product=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.barplot(x='Product_Category',y='Amount',data=sales_product)
```

From the above graph we can conclude that most of the products sold are from Food,Clothing and ELectronics category

```python
sales_product=df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.barplot(x='Product_ID',y='Orders',data=sales_product)
```

Married women belonging to age group 26-35 yrs from UP,Maharastra and Karnataka working in IT,Healthcare and Aviation are more likely oy buy products from Food,clothing and Electronics category
